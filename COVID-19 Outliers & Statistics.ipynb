```python
# Read .csv
conf_hospitalized_state = pd.read_csv("../Project-1/COVID19Datasets/Coronavirus_Numbers_By_State.csv",thousands=',').fillna(0)
```

Using the .csv above, I wanted to exclude using a .iloc function any columns that were irrelevant to the data exploration for this section. I wanted to look at State vs. Positive cases and patients that were hospitalized currently and cumulatively.

I created a new DataFrame in the cell below, using a .loc function, to specifically pull data from the states that had the highest current hospitalization rate (greater than 1000). <b><font size="4">Confirmed Cases vs. Deaths - Among Countries</font></b>

```python
# Read .csv
cases_by_countries = pd.read_csv("../Project-1/COVID19Datasets/COVID-19_Worldwide.csv",thousands=',').fillna(0)
```

Using the .csv above, I wanted to exclude using a .iloc function any columns that were irrelevant to the data exploration for this section. I wanted to analyze confirmed cases vs. deaths among various countries.

```python
# Clean data and build DataFrame
cases_by_countries.iloc[:, [2,3,4,5,6,8,9]].head()
```

<b><font size=3>How many COVID-19 deaths have been confirmed? </font></b>

I plotted the number of cases vs. deaths to analyze the number of COVID-19 deaths confirmed using an x/y logarithm. I used the sum of cases and deaths and set two conditions for cases and deaths (greater than 0). I used a mask function to avoid any non-positive values in x or y. 

```python
# Plot confirmed cases vs. deaths based on countries
total_cases_deaths = cases_by_countries.groupby('countriesAndTerritories')[['cases', 'deaths']].sum()
mask = (total_cases_deaths['cases'] > 0) & (total_cases_deaths['deaths'] > 0)
total_cases_deaths[mask].plot.scatter(x='cases', y='deaths', xlim=(1, 1e7), ylim=(0.5, 1e6), loglog=True, grid=True)
plt.title("Confirmed Cases vs. Deaths")
```

I created a new DataFrame in the cell below to analyze the average and max values for cases and deaths. I set the DataFrame index using existing columns.

```python
# Build a DataFrame to compute the average and max values for cases and deaths
pd.DataFrame(data={'cases':[total_cases_deaths['cases'].mean(), total_cases_deaths['cases'].max()],
                   'deaths':[total_cases_deaths['deaths'].mean(), total_cases_deaths['deaths'].max()]},
             index=['mean', 'max'])
```

The DataFrame presented various outliers so I chose to compare Italy and the U.S based on the statistics in the 'deaths' column. The U.S. shows a high rate in progression of COVID-19 deaths, whereas Italy peaked late March and the death rate declined in April. The dataset was too large to plot deaths based on every country and would have skewed the results.

```python
# Plot COVID-19 deaths per Italy and U.S.
fig, ax = plt.subplots()
countries = ['Italy', 'United_States_of_America']
cases_by_countries['date'] = pd.to_datetime(cases_by_countries['dateRep'])
mask = cases_by_countries['countriesAndTerritories'].isin(countries)
for name, grp in cases_by_countries[mask].groupby('countriesAndTerritories'):
    grp.plot(ax=ax, kind='line', x='date', y='deaths', label=name)
plt.title('COVID-19 Deaths - Italy and U.S.')
plt.ylabel('Deaths')
plt.show()
```

<b>Major Findings</b>
The death rate in the U.S. and Italy does not peak until March. From January to March, the death rate remained at 0. Again, we see that the average number of deaths was significantly higher compared to the average number of cases. The max value of cases was also higher than the max value of deaths. This further confirms that the number of cases have no correlation with the number of deaths. 

The logarithmic function plotting cases vs. deaths suggested a positive correlation between x and y. Death counts rose as the number of confirmed cases increased. Given that this is calculating the sum of cases vs. deaths, the findings of a positive correlation support the data. 

<b><font size = 3>What are the statistics of confirmed COVID-19 cases, deaths and percentage of cases based on population data? </font></b>

I wanted to analyze the cases vs. death counts using a .groupby function to group data by a series of columns. Next, I performed a .sum() function to analyze the sum of these columns. 

```python
# Build a DataFrame to group by countries vs. cases and deaths
case_and_death_counts = cases_by_countries.groupby('countriesAndTerritories')[['cases', 'deaths']].sum()
case_and_death_counts.head()
```

I created a new DataFrame in the cell below to analyze the population count for each country. To analyze, I pulled the mean, using a .mean() function, to analyze the percentage of cases based on population.

```python
# Clean data and build DataFrame
country_pops = cases_by_countries.groupby('countriesAndTerritories')['popData2018'].mean().to_frame(name = 'popData2018').reset_index()
country_pops.head()
```

I created a new variable 'pop_stats' to merge the DataFrame above and set a condition to clean the dataset (greater than 0). Then, I calculated the percentage of cases based on country by dividing the number of cases among each country by population data and multiplying by 100 to achieve the percentages in column 'case %'. The dataset was too large to examine as a whole, as a result a conditional value was set (greater than 0.37). 

```python
# Calculate percentage of cases based on population
pop_stats = country_pops.merge(case_and_death_counts, on='countriesAndTerritories')
cleaned_pop_stats = pop_stats.loc[pop_stats['popData2018'] > 0].copy()
cleaned_pop_stats['case %'] = cleaned_pop_stats['cases'] / cleaned_pop_stats['popData2018'] * 100
cleaned_pop_stats.loc[cleaned_pop_stats['case %'] > 0.37]
```

<b>Major Findings</b>
Most countries and territories, when compared against population data, had less than 1% of cases. This could be viewed as an analomy as the dataset calculated deviated significantly and did not show an even pattern, compared to the previous analyses above.  

The sum of cases among countries was significantly higher, compared to the mortality rate. We could pre-analyze these findings to predict that contracting the virus did not have a correlation with the rate of deaths (per country basis).
